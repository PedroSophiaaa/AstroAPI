```python
import requests
import json
import pandas as pd
import numpy as np
from datetime import date
```

```python
#Definition of latmax and latmin
latmax = 90.0
latmin = -90.0
lista_lat = np.linspace(latmax, latmin, 250)
lon = 20.0 #Fixed
```

```python
api_key = '62c7be520115fd4bbab9147af8d7105a' #Your API key
```

```python
link = 'http://api.weatherstack.com/current' #API link
```

CÃ³digo para salvar os arquivos em .json

```python
for lat in range(len(lista_lat)):
    
    params = {
        'access_key': f'{api_key}',
        'query': f'{lista_lat[lat]},{lon}',
        'units': 'm',
    }
    
    request = requests.get(link, params)

    if str(request) == '<Response [200]>':
        print('ÃŠxito ðŸ‘Œ')
    else:
        print('#Error# âŒ')
        print(link)
        
    json_obj = request.json()
    
    print(json.dumps(json_obj, indent=2))
    print('\n\n\n')
    
    with open(f'data-{lat}-({str(lista_lat[lat]).replace(".", "_")}).json', 'w', encoding='utf-8') as f:
        json.dump(json_obj, f, ensure_ascii=False, indent=4)
```
